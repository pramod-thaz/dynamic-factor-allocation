```python
import yfinance as yf
import pandas as pd
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
#========================== 1. DATA DOWNLOAD ==========================
tickers = ['SPY', 'VTV', 'MTUM', 'QUAL', 'USMV', 'VUG', 'IJR', 'TLT', 'GLD']
print("Downloading ETF data...")
data = yf.download(tickers, start='2013-01-01', end=datetime.today().strftime('%Y-%m-%d'))['Close']
returns = data.pct_change().dropna()
print(f"Data shape: {returns.shape} (daily returns from {returns.index[0].date()} to {returns.index[-1].date()})")
```

```python
# ========================== 2. FEATURE ENGINEERING (Regime Proxies) ==========================
features = pd.DataFrame(index=returns.index)
for t in tickers:
    features[f'{t}_ret'] = returns[t]
    features[f'{t}_vol20'] = returns[t].rolling(20).std()
    features[f'{t}_mom60'] = returns[t].rolling(60).mean()

# ETF-derived regime proxies (these make the model regime-aware)
features['value_mom_spread'] = features['VTV_ret'] - features['MTUM_ret']
features['lowvol_market_spread'] = features['USMV_ret'] - features['SPY_ret']
features['growth_value_spread'] = features['VUG_ret'] - features['VTV_ret']
features['bond_gold_spread'] = features['TLT_ret'] - features['GLD_ret']
features['small_market_spread'] = features['IJR_ret'] - features['SPY_ret']

# Drop NaNs
features = features.dropna()
```

```python
# ========================== 3. SIMPLE REGIME DETECTION (walk-forward) ==========================
# We label regimes using SPY volatility and return (3 states: Bull, Bear, High-Vol)
spy_vol = returns['SPY'].rolling(60).std().dropna()
spy_ret = returns['SPY'].rolling(60).mean().dropna()
regime_data = pd.DataFrame({'vol': spy_vol, 'ret': spy_ret}).dropna()

# Simple quantile-based regime (robust and walk-forward compatible)
regime_data['regime'] = 0  # 0 = Normal, 1 = Bull, 2 = Bear/High-Vol
regime_data.loc[regime_data['ret'] > regime_data['ret'].quantile(0.7), 'regime'] = 1   # Bull
regime_data.loc[(regime_data['vol'] > regime_data['vol'].quantile(0.8)) | 
                (regime_data['ret'] < regime_data['ret'].quantile(0.2)), 'regime'] = 2  # Bear/High-Vol

features = features.join(regime_data['regime'], how='inner')
print("Regimes identified (0=Normal, 1=Bull, 2=Bear/High-Vol):")
print(features['regime'].value_counts(normalize=True))
```

```python
def run_gp_backtest(risk_aversion=1.0, rebal_freq_months=1):
    # Use monthly data (much more stable)
    monthly_returns = returns.resample('ME').agg(lambda x: (1 + x).prod() - 1)
    monthly_features = features.resample('ME').last()
    
    portfolio_returns = []
    weights_history = []
    dates = []
    
    rebal_dates = monthly_features.index[::rebal_freq_months]

    for i in range(1, len(rebal_dates)):
        train_end = rebal_dates[i-1]
        test_date = rebal_dates[i]
        
        train = monthly_features.loc[:train_end]
        X_train = train.drop(columns=['regime']).values.astype(float)
        y_train = {t: monthly_returns.loc[:train_end, t].values.astype(float) for t in tickers}
        
        # Ensure test_row is always (1, n_features)
        test_row = monthly_features.loc[test_date:test_date].drop(columns=['regime']).values.astype(float)
        test_row = test_row.reshape(1, -1)   # Force correct shape
        
        preds_mean = {}
        preds_std = {}
        
        for t in tickers:
            with pm.Model() as gp_model:
                cov_func = pm.gp.cov.Matern52(input_dim=X_train.shape[1], ls=1.0)
                mean_func = pm.gp.mean.Zero()
                gp = pm.gp.Marginal(mean_func=mean_func, cov_func=cov_func)
                
                sigma = pm.HalfCauchy("sigma", beta=0.05)
                jitter = 1e-5
                
                # Correct PyMC 5 syntax
                f = gp.marginal_likelihood(f"f_{t}", X_train, y_train[t], sigma + jitter)
                
                obs = pm.StudentT("obs", mu=f, nu=5, sigma=sigma, observed=y_train[t])
                
                trace = pm.sample(300, tune=200, chains=2, target_accept=0.98,
                                  init='jitter+adapt_diag', progressbar=False, random_seed=42)
            
            # Posterior predictive with explicit shape handling
            with gp_model:
                f_pred = gp.conditional(f"f_pred_{t}", Xnew=test_row)
                post_pred = pm.sample_posterior_predictive(
                    trace, var_names=[f"f_pred_{t}"], progressbar=False
                )
            
            pred_samples = post_pred.posterior_predictive[f"f_pred_{t}"].values
            preds_mean[t] = float(pred_samples.mean())
            preds_std[t] = float(pred_samples.std())
        
        # Portfolio construction
        mu = np.array([preds_mean[t] for t in tickers])
        sigma_vec = np.array([preds_std[t] for t in tickers])
        cov = np.diag(sigma_vec**2 + 1e-6)
        
        def objective(w):
            port_ret = w @ mu
            port_vol = np.sqrt(w.T @ cov @ w)
            return -(port_ret - risk_aversion * port_vol)
        
        cons = {'type': 'eq', 'fun': lambda w: np.sum(w) - 1}
        bounds = [(0, 1)] * len(tickers)
        res = minimize(objective, np.ones(len(tickers))/len(tickers), 
                       bounds=bounds, constraints=cons, method='SLSQP')
        weights = res.x
        
        # Realized monthly return
        port_ret = (weights * monthly_returns.loc[test_date, tickers]).sum()
        
        portfolio_returns.append(port_ret)
        weights_history.append(weights)
        dates.append(test_date)
    
    backtest_df = pd.DataFrame({
        'date': dates,
        'return': portfolio_returns,
        'cum_ret': np.cumprod(1 + np.array(portfolio_returns)) - 1
    }).set_index('date')
    
    return backtest_df, weights_history, preds_mean, preds_std
```

```python
# ========================== 5. OUTER HSGP SURROGATE (Bayesian Optimization) ==========================
# Simple outer loop: optimize risk_aversion (you can add rebal_freq etc.)
print("Running outer HSGP surrogate optimization (small grid for demo)...")
risk_av_values = np.linspace(0.5, 2.0, 8)
results = []
for ra in risk_av_values:
    print(f"Testing risk_aversion = {ra:.2f}")
    bt, _, _, _ = run_gp_backtest(risk_aversion=ra)
    sharpe = bt['return'].mean() / bt['return'].std() * np.sqrt(252) if len(bt) > 0 else 0
    results.append((ra, sharpe, bt['cum_ret'].iloc[-1]))

best_ra = max(results, key=lambda x: x[1])[0]
print(f"Best risk aversion found: {best_ra:.2f}")

# Run final backtest with best hyperparams
final_bt, weights_hist, pred_means, pred_stds = run_gp_backtest(risk_aversion=best_ra)
```
